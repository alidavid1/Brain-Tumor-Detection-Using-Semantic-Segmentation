## Goal

The aim of this practice is to apply deep learning techniques for the semantic segmentation of brain tumors in medical imaging using  Magnetic Resonance Imaging (MRI) data from a public dataset such as BRATS, available on platforms like Kaggle. The student will implement and train a convolutional neural network for segmentation, apply appropriate pre- and post-processing steps, and evaluate the model using metrics specific to segmentation performance. The student must use PyTorch as the deep learning framework, OpenCV for image processing, and matplotlib or similar libraries for the visual analysis of results.


## General description

The pipeline to be developed should follow these stages:

1. Load and preprocess the MRI dataset from Kaggle.
2. Split the data into training, validation and test sets.
3. Resize and normalize the MRI slices.
4. Prepare the corresponding binary segmentation masks.
5. Design and implement a semantic segmentation network, preferably U-Net or a similar encoder-decoder architecture.
6. Train the model using appropriate loss functions for segmentation, such as Binary Cross-Entropy, Dice Loss or a combination of both.
7. Validate the model on unseen data and visualize predictions.
8. Evaluate performance using segmentation metrics.
9. Provide qualitative results showing predicted masks overlaid on the original MRI slices.

Concepts related to this practice have been studied mainly in Topic 5 of the course. It is recommended that the practical implementation is carried out using the previous practices as a basis, since they may serve as a guide for implementing the full segmentation pipeline.

The data for the training and evaluation of the algorithm will be obtained from the [BraTS 2020 Training Data](https://www.kaggle.com/datasets/awsaf49/brats2020-training-data) available on Kaggle. This dataset contains pre-operative MRI scans of brain tumor patients, collected from multiple institutions and annotated by expert neuroradiologists. Each subject includes four MRI modalities: T1-weighted, T1 with contrast enhancement, T2-weighted and FLAIR, together with ground-truth segmentation masks identifying different tumor sub-regions.

In this practice, students must use the FLAIR modality and convert the segmentation labels into a binary tumor mask. The tumor mask must be obtained by combining the available tumor labels into a single foreground class. Semantic segmentation is critical in this context to accurately localize tumor areas that may be difficult to distinguish visually due to low contrast, intensity inhomogeneity and irregular boundaries. An example from the BraTS dataset is shown in Figure 1, illustrating the multimodal MRI input and corresponding multi-class segmentation masks.

The dataset consists of 369 subjects. It is recommended to follow a 70% training, 15% validation and 15% test split to ensure consistent model evaluation.

To perform semantic segmentation on the MRI slices, a convolutional neural network architecture based on U-Net will be used. U-Net is a widely adopted encoder-decoder architecture designed for biomedical image segmentation tasks. Its structure allows precise localization through skip connections that combine high-resolution features from the contracting path with the upsampled output. This facilitates the segmentation of small or irregular tumor regions in brain MRI images. To evaluate the performance of the segmentation model, standard segmentation metrics must be used. Specifically, the Dice coefficient and the Intersection over Union, IoU, are used to measure the overlap between the predicted mask and the ground-truth mask. These metrics are particularly suitable for medical image segmentation, where region overlap is more informative than global pixel accuracy.

The Dice coefficient is defined as:

$$
\text{Dice} = \frac{2 \cdot \lvert A \cap B \rvert}{\lvert A \rvert + \lvert B \rvert}
$$

The Intersection over Union, IoU, is defined as:

$$
\text{IoU} = \frac{\lvert A \cap B \rvert}{\lvert A \cup B \rvert}
$$

where *A* is the predicted segmentation mask and *B* is the ground-truth mask.
