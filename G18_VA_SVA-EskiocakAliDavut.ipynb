```python
import os
import glob
import random
import numpy as np
import matplotlib.pyplot as plt
import h5py  # <-- ADDED for .h5 files
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import albumentations as A
from albumentations.pytorch import ToTensorV2
from sklearn.model_selection import train_test_split
from tqdm import tqdm

def seed_everything(seed=42):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True

seed_everything(42)

# Configuration
CONFIG = {
    # Updated path based on your Kaggle screenshot
    "DATA_PATH": "/Users/alidavuteskiocak/.cache/kagglehub/datasets/awsaf49/brats2020-training-data/versions/3/BraTS2020_training_data/content/data",
    "IMAGE_SIZE": 128,
    "BATCH_SIZE": 32,
    "EPOCHS": 15,
    "LEARNING_RATE": 1e-4,
    "DEVICE": "cuda" if torch.cuda.is_available() else "cpu"
}
```

```python
def get_patient_data(base_path):
    # Find all .h5 slice files directly in the directory
    h5_files = glob.glob(os.path.join(base_path, "*.h5"))
    return h5_files

class BraTSDataset(Dataset):
    def __init__(self, data_list, transform=None):
        self.data_list = data_list
        self.transform = transform

    def __len__(self):
        return len(self.data_list)

    def __getitem__(self, idx):
        file_path = self.data_list[idx]
        
        # Read the .h5 file
        with h5py.File(file_path, 'r') as f:
            image = np.array(f['image'])
            mask = np.array(f['mask'])
            
        # 1. Extract FLAIR channel
        # Standard .h5 BraTS sets have shape (240, 240, 4) -> [FLAIR, T1w, T1gd, T2w]
        if image.ndim == 3 and image.shape[-1] == 4:
            flair_img = image[:, :, 0] 
        elif image.ndim == 3 and image.shape[0] == 4:
            flair_img = image[0, :, :]
        else:
            flair_img = image 
            
        # 2. Combine multi-class mask into binary mask
        if mask.ndim == 3:
            mask_img = np.max(mask, axis=-1)
        else:
            mask_img = mask
            
        mask_img[mask_img > 0] = 1.0
        
        flair_img = np.array(flair_img, dtype=np.float32)
        mask_img = np.array(mask_img, dtype=np.float32)
        
        # Normalize FLAIR
        if np.max(flair_img) > 0:
            flair_img = flair_img / np.max(flair_img)
            
        if self.transform:
            augmented = self.transform(image=flair_img, mask=mask_img)
            flair_img = augmented['image']
            mask_img = augmented['mask']
        else:
            flair_img = torch.from_numpy(flair_img).unsqueeze(0)
            mask_img = torch.from_numpy(mask_img).unsqueeze(0)
            
        return flair_img, mask_img

# Splitting data 70/15/15
all_data = get_patient_data(CONFIG["DATA_PATH"])

# Safety check so we don't fail silently again
if len(all_data) == 0:
    raise ValueError(f"No .h5 files found in {CONFIG['DATA_PATH']}. Please double-check the path using the Kaggle file explorer on the right!")

train_data, temp_data = train_test_split(all_data, test_size=0.3, random_state=42)
val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

print(f"Train slices: {len(train_data)} | Val slices: {len(val_data)} | Test slices: {len(test_data)}")

# Data Augmentation
train_transform = A.Compose([
    A.Resize(CONFIG["IMAGE_SIZE"], CONFIG["IMAGE_SIZE"]),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.0625, scale_limit=0.1, rotate_limit=45, p=0.5),
    ToTensorV2()
])

val_test_transform = A.Compose([
    A.Resize(CONFIG["IMAGE_SIZE"], CONFIG["IMAGE_SIZE"]),
    ToTensorV2()
])

train_loader = DataLoader(BraTSDataset(train_data, train_transform), batch_size=CONFIG["BATCH_SIZE"], shuffle=True, num_workers=2)
val_loader = DataLoader(BraTSDataset(val_data, val_test_transform), batch_size=CONFIG["BATCH_SIZE"], shuffle=False, num_workers=2)
test_loader = DataLoader(BraTSDataset(test_data, val_test_transform), batch_size=CONFIG["BATCH_SIZE"], shuffle=False, num_workers=2)
```

```python
class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        return self.conv(x)

class UNet(nn.Module):
    def __init__(self, in_channels=1, out_channels=1, features=[64, 128, 256, 512]):
        super(UNet, self).__init__()
        self.downs = nn.ModuleList()
        self.ups = nn.ModuleList()
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # Encoder
        for feature in features:
            self.downs.append(DoubleConv(in_channels, feature))
            in_channels = feature

        # Bottleneck
        self.bottleneck = DoubleConv(features[-1], features[-1]*2)

        # Decoder
        for feature in reversed(features):
            self.ups.append(
                nn.ConvTranspose2d(feature*2, feature, kernel_size=2, stride=2)
            )
            self.ups.append(DoubleConv(feature*2, feature))

        self.final_conv = nn.Conv2d(features[0], out_channels, kernel_size=1)

    def forward(self, x):
        skip_connections = []

        for down in self.downs:
            x = down(x)
            skip_connections.append(x)
            x = self.pool(x)

        x = self.bottleneck(x)
        skip_connections = skip_connections[::-1]

        for idx in range(0, len(self.ups), 2):
            x = self.ups[idx](x)
            skip_connection = skip_connections[idx//2]
            
            # Handle resizing if dimensions don't match perfectly
            if x.shape != skip_connection.shape:
                x = nn.functional.interpolate(x, size=skip_connection.shape[2:], mode="bilinear", align_corners=True)
                
            concat_skip = torch.cat((skip_connection, x), dim=1)
            x = self.ups[idx+1](concat_skip)

        return torch.sigmoid(self.final_conv(x))

model = UNet(in_channels=1, out_channels=1).to(CONFIG["DEVICE"])
```

```python
class BCEDiceLoss(nn.Module):
    def __init__(self):
        super(BCEDiceLoss, self).__init__()
        self.bce = nn.BCELoss()

    def forward(self, inputs, targets, smooth=1e-6):
        inputs = inputs.view(-1)
        targets = targets.view(-1)
        
        # BCE
        bce_loss = self.bce(inputs, targets)
        
        # Dice
        intersection = (inputs * targets).sum()
        dice_loss = 1 - (2. * intersection + smooth) / (inputs.sum() + targets.sum() + smooth)
        
        return bce_loss + dice_loss

def calculate_metrics(pred, target, smooth=1e-6):
    pred = (pred > 0.5).float()
    
    intersection = (pred * target).sum(dim=(1, 2, 3))
    union = pred.sum(dim=(1, 2, 3)) + target.sum(dim=(1, 2, 3)) - intersection
    
    iou = (intersection + smooth) / (union + smooth)
    dice = (2. * intersection + smooth) / (pred.sum(dim=(1, 2, 3)) + target.sum(dim=(1, 2, 3)) + smooth)
    
    return iou.mean().item(), dice.mean().item()

criterion = BCEDiceLoss()
optimizer = optim.Adam(model.parameters(), lr=CONFIG["LEARNING_RATE"])
```

```python
train_losses, val_losses = [], []
train_dices, val_dices = [], []

for epoch in range(CONFIG["EPOCHS"]):
    # Training Phase
    model.train()
    running_loss, running_dice = 0.0, 0.0
    
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['EPOCHS']} - Train")
    for images, masks in loop:
        images, masks = images.to(CONFIG["DEVICE"]).float(), masks.to(CONFIG["DEVICE"]).float()
        
        # Since transforms might strip the channel dimension:
        if images.dim() == 3: images = images.unsqueeze(1)
        if masks.dim() == 3: masks = masks.unsqueeze(1)
            
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()
        
        _, dice = calculate_metrics(outputs, masks)
        running_loss += loss.item()
        running_dice += dice
        loop.set_postfix(loss=loss.item(), dice=dice)
        
    train_losses.append(running_loss / len(train_loader))
    train_dices.append(running_dice / len(train_loader))
    
    # Validation Phase
    model.eval()
    val_loss, val_dice = 0.0, 0.0
    with torch.no_grad():
        loop = tqdm(val_loader, desc=f"Epoch {epoch+1}/{CONFIG['EPOCHS']} - Val")
        for images, masks in loop:
            images, masks = images.to(CONFIG["DEVICE"]).float(), masks.to(CONFIG["DEVICE"]).float()
            if images.dim() == 3: images = images.unsqueeze(1)
            if masks.dim() == 3: masks = masks.unsqueeze(1)

            outputs = model(images)
            loss = criterion(outputs, masks)
            _, dice = calculate_metrics(outputs, masks)
            
            val_loss += loss.item()
            val_dice += dice
            loop.set_postfix(loss=loss.item(), dice=dice)
            
    val_losses.append(val_loss / len(val_loader))
    val_dices.append(val_dice / len(val_loader))

# Plot Learning Curves
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.title('Loss over Epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_dices, label='Train Dice')
plt.plot(val_dices, label='Val Dice')
plt.title('Dice Score over Epochs')
plt.legend()
plt.show()

# Save the final model weights
torch.save(model.state_dict(), "unet_brats.pth")
```

```python
model.eval()
test_ious, test_dices = [], []

with torch.no_grad():
    for images, masks in tqdm(test_loader, desc="Testing"):
        images, masks = images.to(CONFIG["DEVICE"]).float(), masks.to(CONFIG["DEVICE"]).float()
        if images.dim() == 3: images = images.unsqueeze(1)
        if masks.dim() == 3: masks = masks.unsqueeze(1)
            
        outputs = model(images)
        
        # Calculate per-batch metrics, iterating to get per-image metrics
        preds = (outputs > 0.5).float()
        for p, t in zip(preds, masks):
            intersection = (p * t).sum()
            union = p.sum() + t.sum() - intersection
            
            iou = (intersection + 1e-6) / (union + 1e-6)
            dice = (2. * intersection + 1e-6) / (p.sum() + t.sum() + 1e-6)
            
            test_ious.append(iou.item())
            test_dices.append(dice.item())

print(f"Test IoU:  {np.mean(test_ious):.4f} ± {np.std(test_ious):.4f}")
print(f"Test Dice: {np.mean(test_dices):.4f} ± {np.std(test_dices):.4f}")

# Boxplot for variance analysis
plt.figure(figsize=(6, 4))
plt.boxplot([test_ious, test_dices], labels=['IoU', 'Dice'])
plt.title('Test Set Metrics Distribution')
plt.ylabel('Score')
plt.show()
```

```python
def visualize_predictions(dataset, num_samples=3):
    model.eval()
    fig, axes = plt.subplots(num_samples, 4, figsize=(15, 4*num_samples))
    axes[0, 0].set_title("Input FLAIR", fontsize=14)
    axes[0, 1].set_title("Ground Truth", fontsize=14)
    axes[0, 2].set_title("Prediction", fontsize=14)
    axes[0, 3].set_title("Overlay (GT:Green, Pred:Red)", fontsize=14)
    
    with torch.no_grad():
        for i in range(num_samples):
            # Pick random samples
            idx = random.randint(0, len(dataset) - 1)
            image, mask = dataset[idx]
            
            # Prepare image for model
            img_tensor = image.unsqueeze(0).to(CONFIG["DEVICE"]).float()
            if img_tensor.dim() == 3: img_tensor = img_tensor.unsqueeze(1)
                
            pred = model(img_tensor)
            pred = (pred > 0.5).squeeze().cpu().numpy()
            
            img_np = image.squeeze().numpy()
            mask_np = mask.squeeze().numpy()
            
            # Create RGB overlay
            overlay = np.stack((img_np, img_np, img_np), axis=-1)
            # Add ground truth in Green
            overlay[mask_np == 1] = [0, 1, 0]
            # Add prediction in Red
            overlay[pred == 1] = [1, 0, 0]
            # Intersection becomes Yellow ([1,1,0])
            overlay[(mask_np == 1) & (pred == 1)] = [1, 1, 0]
            
            axes[i, 0].imshow(img_np, cmap='gray')
            axes[i, 0].axis('off')
            
            axes[i, 1].imshow(mask_np, cmap='gray')
            axes[i, 1].axis('off')
            
            axes[i, 2].imshow(pred, cmap='gray')
            axes[i, 2].axis('off')
            
            axes[i, 3].imshow(overlay)
            axes[i, 3].axis('off')
            
    plt.tight_layout()
    plt.show()

print("Displaying qualitative results on the Test set...")
visualize_predictions(BraTSDataset(test_data, val_test_transform), num_samples=5)
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from torch.utils.data import DataLoader

# --- ÇÖZÜM BURADA: Multiprocessing Hatasını Aşmak ---
# Mevcut test_loader'ın dataset'ini alıp, num_workers=0 ile güvenli bir loader oluşturuyoruz.
# Bu sayede Mac/Jupyter üzerindeki "BrokenPipe" ve "AttributeError" hatalarını atlıyoruz.
safe_test_loader = DataLoader(
    test_loader.dataset, 
    batch_size=test_loader.batch_size, 
    shuffle=False, 
    num_workers=0  # <--- Hatanın çözümü bu parametre
)

# 1. Modeli başlat ve eğitilmiş ağırlıkları yükle
worst_case_model = UNet(in_channels=1, out_channels=1).to(CONFIG["DEVICE"])
worst_case_model.load_state_dict(torch.load("unet_brats.pth", map_location=CONFIG["DEVICE"]))
worst_case_model.eval()

results = []

print("Test seti değerlendiriliyor ve en kötü skorlar aranıyor...")

# 2. Test seti üzerinde tahminleri yap ve skorları kaydet
with torch.no_grad():
    # Burada safe_test_loader kullanıyoruz
    for images, masks in tqdm(safe_test_loader, desc="Evaluating Worst Cases"):
        images = images.to(CONFIG["DEVICE"]).float()
        masks = masks.to(CONFIG["DEVICE"]).float()
        
        if images.dim() == 3: images = images.unsqueeze(1)
        if masks.dim() == 3: masks = masks.unsqueeze(1)
            
        outputs = worst_case_model(images)
        preds = (outputs > 0.5).float()
        
        # Batch içindeki her bir görüntüyü tek tek analiz et
        for i in range(images.size(0)):
            p = preds[i]
            t = masks[i]
            
            # Dice Skoru Hesaplama
            intersection = (p * t).sum()
            dice = (2. * intersection + 1e-6) / (p.sum() + t.sum() + 1e-6)
            
            # Sonuçları CPU'ya çekip listeye ekle
            results.append({
                'dice': dice.item(),
                'image': images[i].cpu().squeeze().numpy(),
                'mask': masks[i].cpu().squeeze().numpy(),
                'pred': preds[i].cpu().squeeze().numpy()
            })

# 3. Sonuçları Dice skoruna göre küçükten büyüğe (en kötüden en iyiye) sırala
results.sort(key=lambda x: x['dice'])

# En kötü 3 durumu al
worst_3_cases = results[:3]

print("\nEn kötü 3 tahmin görselleştiriliyor...")

# 4. Görselleştirme
fig, axes = plt.subplots(3, 4, figsize=(15, 12))
axes[0, 0].set_title("Input FLAIR", fontsize=14)
axes[0, 1].set_title("Ground Truth", fontsize=14)
axes[0, 2].set_title("Prediction", fontsize=14)
axes[0, 3].set_title("Overlay (GT:Green, Pred:Red)", fontsize=14)

for i, case in enumerate(worst_3_cases):
    img_np = case['image']
    mask_np = case['mask']
    pred_np = case['pred']
    score = case['dice']
    
    # Overlay oluşturma
    overlay = np.stack((img_np, img_np, img_np), axis=-1)
    overlay[mask_np == 1] = [0, 1, 0] # GT: Yeşil
    overlay[pred_np == 1] = [1, 0, 0] # Pred: Kırmızı
    overlay[(mask_np == 1) & (pred_np == 1)] = [1, 1, 0] # Kesişim: Sarı
    
    # Çizimler
    axes[i, 0].imshow(img_np, cmap='gray')
    axes[i, 0].axis('off')
    axes[i, 0].text(0, -10, f"Worst Case #{i+1}", color='red', fontsize=12, fontweight='bold')
    
    axes[i, 1].imshow(mask_np, cmap='gray')
    axes[i, 1].axis('off')
    
    axes[i, 2].imshow(pred_np, cmap='gray')
    axes[i, 2].axis('off')
    axes[i, 2].text(0, -10, f"Dice: {score:.4f}", color='red', fontsize=12, fontweight='bold')
    
    axes[i, 3].imshow(overlay)
    axes[i, 3].axis('off')

plt.tight_layout()
plt.show()
```

# Worst Case Analysis: Catastrophic Prediction Failure

## 1. Visual Observations
Upon analyzing the worst-performing cases from the test set, a clear and consistent failure mode emerges across all three instances:

Input (FLAIR): The input MRIs contain clearly visible anatomical structures and hyperintense regions indicative of anomalies.

Ground Truth: The ground truth masks contain large, distinct, and contiguous tumor regions. These are not small, ambiguous micro-lesions; they are prominent macroscopic structures.

Prediction: The model's prediction masks are entirely black. The network failed to predict a single foreground (tumor) pixel.

Metric: Consequently, the Dice Score for all these cases is exactly 0.0000. The overlay confirms that the model completely missed the regions of interest (ROI).

## 2. Root Cause Analysis (Why is the model struggling?)
The fact that the model predicts a completely blank mask for prominent tumors indicates a systemic training failure rather than a lack of model capacity to capture fine details. The model has suffered from a phenomenon often referred to as "All-Background Prediction" or "Dying Model Syndrome". The primary suspects for this behavior are:

Severe Class Imbalance: In brain MRI slices, the vast majority of pixels belong to the background (empty space or healthy tissue), while the tumor occupies only a small fraction. If the loss function does not sufficiently penalize false negatives, the model learns the "lazy" solution: predicting $0$ (background) for every pixel. This strategy yields a low Binary Cross Entropy (BCE) loss and high pixel-wise accuracy, tricking the optimizer into a local minimum early in training.

Thresholding Bottleneck: The final output of the network passes through a Sigmoid activation, yielding probabilities between $0$ and $1$. We apply a hard threshold of $> 0.5$ to generate the binary mask. It is highly probable that the model is outputting very weak probabilities (e.g., $0.1$ or $0.3$ in the tumor regions) that never breach the $0.5$ threshold, resulting in an entirely zeroed-out array after binarization.

Suboptimal Learning Rate / Dead Gradients: As observed in earlier hyperparameter tuning (e.g., $LR=1e-4$ vs $1e-3$), the model might be taking steps that are either too small to escape the initial all-background local minimum, or too large, causing the gradients to collapse and the ReLUs to "die" (outputting zero consistently).

## 3. Potential Mitigations & Next Steps
To prevent the model from defaulting to blank predictions, the following strategies should be investigated:

Loss Function Tuning: While BCEDiceLoss is currently implemented, the BCE component might be dominating the Dice component in the early epochs. Applying class weights to the BCE (penalizing tumor misses more heavily) or relying entirely on Dice Loss / Focal Loss could force the network to focus on the foreground.

Threshold Adjustment: Instead of a strict $0.5$ threshold, analyzing the histogram of the raw Sigmoid outputs could reveal if the network is detecting the tumor but with low confidence. Lowering the threshold (e.g., to $0.3$) might recover some of these masks.

Data Normalization Verification: Ensure that the input FLAIR images are strictly normalized (e.g., z-score normalization or Min-Max scaling to $[0, 1]$). Extreme outlier pixel values in unnormalized MRI data can cause catastrophic gradient updates.

# Comparison of Training Behavior: With vs. Without Data Augmentation

When training a model like U-Net on medical imaging data (such as the BraTS dataset), the presence or absence of data augmentation fundamentally changes the learning dynamics. Here is a detailed breakdown of how the training behavior shifts and how it impacts generalization.

## 1. Training Without Data Augmentation

Without spatial transformations, the model sees the exact same pixel-perfect images in every single epoch. 

* **Training Metrics:** The training loss will decrease very rapidly, and the training Dice score will quickly approach 1.0 (perfect overlap). The model easily memorizes the exact position, size, and orientation of the tumors in the training set.
* **Validation Metrics:** The validation loss will initially drop but will likely plateau early (around epoch 5-8) and then start to increase. The validation Dice score will stagnate.
* **The "Gap":** You will observe a massive gap between your training and validation curves. For example, your Train Dice might hit 0.96, while your Val Dice gets stuck at 0.78. 
* **Diagnosis:** **Severe Overfitting**. The model has learned the specific training images rather than the general features of a brain tumor.

## 2. Training With Data Augmentation

By applying your `albumentations` pipeline (Horizontal/Vertical Flips, Rotations, Shift/Scale), the model almost never sees the exact same image twice. 

* **Training Metrics:** The training loss decreases much more slowly and predictably. The training curves might look a bit "noisier" epoch-to-epoch because the difficulty of the data is artificially inflated. The training Dice score will be lower than in the unaugmented run.
* **Validation Metrics:** The validation loss continues to decrease steadily alongside the training loss. 
* **The "Gap":** The training and validation curves will track closely together. You might see a Train Dice of 0.86 and a Val Dice of 0.85. 
* **Diagnosis:** **Healthy Learning**. The model is being regularized by the augmentations, preventing it from relying on exact spatial memorization.

---

## Does Data Augmentation Improve Generalization?

**Yes, significantly—especially in medical imaging.** 

Generalization refers to a model's ability to perform well on unseen data. Data augmentation improves this through several key mechanisms:

1. **Invariance Learning:** A tumor is still a tumor whether it is mirrored, rotated by 90 degrees, or shifted slightly to the left. By artificially creating these variations, you force the U-Net's convolutional filters to learn the *texture*, *shape*, and *intensity* of the tumor rather than its absolute coordinates in the 128x128 grid.
2. **Artificial Dataset Expansion:** Medical datasets are notoriously small due to the cost of expert annotation. Augmentation acts as a multiplier, turning 1,000 unique patient slices into effectively infinite variations, which deep learning models require to optimize their millions of parameters (your U-Net has ~31 million).
3. **Robustness to Clinical Variance:** In real-world inference, patient heads are not perfectly aligned in the MRI scanner. The `ShiftScaleRotate` augmentation mathematically simulates these slight misalignments, meaning the model won't fail if the hospital's MRI tech positions a patient slightly off-center.

**Conclusion:** While removing augmentation makes your *training* graphs look better, keeping augmentation makes your *actual model* better. For complex tasks like segmentation, data augmentation is not just a bonus—it is a strict requirement for achieving a clinically viable model.
